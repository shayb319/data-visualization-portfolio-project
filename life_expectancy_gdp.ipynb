# Import necessary modules

First, we need to import all the essential modules for this project.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Import and inspect data

Now that we've imported the necessary modules, we need to import our data and inspect it before we can continue investigating.

```python
#import file
data = pd.read_csv('all_data.csv')

#print first 5 rows of data
data.head()
```

Our data consists of 4 columns; Country, Year, Life expectancy at birth in years and GDP.

```python
data.info()
```

It seems that we are not missing any values and the current data types are sufficient for our project.

# Explore the data

Let's calculate some summary statistics on our data.

```python
#calculate summary statistics for all variables
data.describe(include = 'all')
```

Some statistics to take note of: <br>
>The average life expectancy for the data is 72.79 years and ranges from 44 to 81<br>
The average GDP for the data is \\$3,880,499,000,000 <br>
The year column ranges from 2000 to 2015 <br>
There are 6 unique countries, each with 16 entries (one for each year)

```python
data.Country.unique()
```

The 6 countries in our data are Chile, China, Germany, Mexico, United States of America and Zimbabwe.

```python
data.Year.unique()
```

This confirms that every year between 2000 and 2015 is included in the dataset.

Let's make a few updates to our data before continuing <br>
First, we'll replace all values of 'United States of America' with 'USA' to make our data and plots more readable.<br>
Next, we'll create seperate dataframes for each country. This will help us during the EDA process.<br>
We'll also create a dictionary of each country name and a corresponding color. This way we'll be able to create charts using the same color for each country throughout our analysis.<br>
Lastly, we're going to rename the life expectancy column  to 'LE' to simplify our future code.

```python
#update all values of 'United States of America' to 'USA'
data.loc[data.Country == 'United States of America', 'Country'] = 'USA'

#rename column
data.rename(columns = {'Life expectancy at birth (years)' : 'LE'}, inplace = True)

#create dataframes for each country
chile_data = data[data.Country == 'Chile']
china_data = data[data.Country == 'China']
germany_data = data[data.Country == 'Germany']
mexico_data = data[data.Country == 'Mexico']
usa_data = data[data.Country == 'USA']
zimbabwe_data = data[data.Country == 'Zimbabwe']

#create a color palette for the countries
country_palette = {'Chile': '#1b9e77', 'China': '#d95f02', 'Germany': '#7570b3', 'Mexico': '#e7298a', 'USA': '#66a61e', 'Zimbabwe': '#e6ab02'}

#print first 5 rows in usa_data to confirm the new dataframes were created and USA values have changed
usa_data.head()
```

## Exploratory Plots

Moving on to some EDA.

First, we'll take a look at the distribution of life expectancy for the entire dataset.

```python
sns.displot(data = data, x = 'LE', height = 8, aspect= 8/8, rug = True)
plt.xlabel('Life expectancy at birth in years')
plt.title('Life Expectancy Distribution')
plt.show()
```

The majority of the data has a life expectancy between 72 and 81, but the data is very left skewed with a larger count around 45 years.

Next, we'll take a look at the distribution of GDP for the entire dataset.

```python
sns.displot(data = data, x = 'GDP', height = 8, aspect = 1, rug = True)
plt.xlabel('GDP in Trillions of U.S. Dollars')
plt.title('GDP Distribution')
plt.show()
```

The majority of the data falls in the lower range of possible GDP values in the data and is very right-skewed.

Now let's take a look at how GDP and Life Expectancy have changed over the years for each country.

```python
plt.figure(figsize=(10,8))
sns.lineplot(data = data, x = 'Year', y = 'GDP', hue = 'Country', palette= country_palette)
plt.ylabel('GDP in Trillions of U.S. Dollars')
plt.title('Yearly GDP from 2000 to 2015')
plt.legend(loc = 2)
plt.show()
```

China and the USA have seen the most growth from 2000 to 2015 in GDP value, but the United States has consistantly had the top value among these countries.<br> Zimbabwe and Chile have the lowest GDP value with little  growth over the years compared to the other countries. 

```python
plt.figure(figsize=(10,8))
sns.lineplot(data = data, x = 'Year', y = 'LE', hue = 'Country', palette= country_palette)
plt.axhline(y = 70, linestyle = '--', linewidth = 1)
plt.title('Yearly Life Expectancy from 2000 to 2015')
plt.ylabel('Life expectancy at birth in years')
plt.show()
```

The life expectancy in Chile, China, Germany, Mexico and the USA have remained over 70 years from 2000 to 2015 with small increases for each country over the years. <br>
Zimbabwe has remained consistantly under 70 years. Starting around 46 years in 2000, they experienced a dip in the early 2000's, but the life expectancy has been increasing consistantly since 2005.

Next, we'll take a look at the average GDP and Life Expectancy for each country.

```python
plt.figure(figsize = (12,8))
sns.barplot(data = data, x = 'Country', y = 'GDP', order = data.groupby('Country')['GDP'].mean().sort_values().index, palette= country_palette)
plt.xticks(rotation = 45)
plt.title('Average GDP by Country')
plt.show()
```

The USA has the largest average GDP of the six nations, more than double the average value of China. <br>
Zimbabwe has the lowest GDP value, but since it is so low, we can't see much of the bar in the plot, therefore, we can't accurately compare it to the other countries. <br>
Instead, let's calculate the values for average GDP by nation so we can see just how much of a difference there is.

```python
#aggregate GDP values by Country, calculate the mean and sort in ascending order
data.groupby('Country')['GDP'].mean().sort_values(ascending = True)
```

Zimbabwe has an average GDP of \\$9 billion, whereas the next highest GDP average value comes from Chile at \\$170 billion. <br>
We mentioned before that the average GDP value of the USA is more than double that of China. Turns out, the USA has almost 3 times the average value at \\$14 trillion while China is around \\$5 trillion.

```python
plt.figure(figsize = (12,8))
sns.violinplot(data = data, x = 'Country', y = 'LE', palette = country_palette, order = data.groupby('Country')['LE'].mean().sort_values().index)
plt.xticks(rotation = 45)
plt.ylabel('Life expectancy at birth in years')
plt.title('Average Life Expectancy by Country')
plt.show()
```

From the above plot, we can see that Zimbabwe falls behind with the lowest average life expectancy at around 50 years, while also having the largest range of values. China, Mexico, the USA, Chile and Germany all have an average life expectancy between 70 and 80 years, with Germany having the highest life expectancy among the six nations. However, it is a little difficult to discern the actual values of the top five nations and how much they differ from each other.

Let's calculate the average life expectancy for each nation.

```python
#aggregate Life Expectancy values by Country, calculate the mean and sort in ascending order
data.groupby('Country')['LE'].mean().sort_values(ascending = True)
```

Recall from earlier that we determined the average life expectancy for the data to be 72.79 years. The calculations we have made above show that 5 out of the 6 nations have an average life expectancy significantly above the average of the entire dataset.<br>
Upon reviewing, we now understand that Zimbabwe is an outlier and had a large impact on the overall mean.

Earlier, we plotted Year vs. GDP for each nation on one chart. Since the yearly GDP could vary greatly from one nation to the next, it was difficult to determine if all of the countries saw an increase in GDP from 2000 to 2015. <br>
Let's plot the yearly GDP for each country and their trend line on different plots.

```python
plt.figure(figsize = (15,15))
plt.subplot(3, 2, 1)
sns.regplot(data = chile_data, x = 'Year', y = 'GDP', color = list(country_palette.values())[0], marker = 'o')
plt.xticks(rotation = 45)
plt.title('Chile GDP by Year')
plt.subplots_adjust(hspace =0.5)

plt.subplot(3, 2, 2)
sns.regplot(data = china_data, x = 'Year', y = 'GDP', color = list(country_palette.values())[1], marker = 'o')
plt.xticks(rotation = 45)
plt.title('China GDP by Year')

plt.subplot(3, 2, 3)
sns.regplot(data = germany_data, x = 'Year', y = 'GDP', color = list(country_palette.values())[2], marker = 'o')
plt.xticks(rotation = 45)
plt.title('Germany GDP by Year')

plt.subplot(3, 2, 4)
sns.regplot(data = mexico_data, x = 'Year', y = 'GDP', color = list(country_palette.values())[3], marker = 'o')
plt.xticks(rotation = 45)
plt.title('Mexico GDP by Year')

plt.subplot(3, 2, 5)
sns.regplot(data = usa_data, x = 'Year', y = 'GDP', color = list(country_palette.values())[4], marker = 'o')
plt.xticks(rotation = 45)
plt.title('USA GDP by Year')

plt.subplot(3, 2, 6)
sns.regplot(data = zimbabwe_data, x = 'Year', y = 'GDP', color = list(country_palette.values())[5], marker = 'o')
plt.xticks(rotation = 45)
plt.title('Zimbabwe GDP by Year')
plt.show()
```

Based on these plots, we can now say with certainty that the GDP in each country increased from the year 2000 to 2015.
Also, notice how between 2007 and 2009, every country except China experienced a dip in their GDP. This was around the time the world went into a recession.

Similarly to our GDP plots, we plotted Life Expectancy by year for each nation on one chart. We could clearly see that the life expectancy in Zimbabwe went up significantly compared to the other five nations, but it's a little more difficult to discern the yearly changes for the other countries and whether there was an increase over time or not. 
Let's plot the yearly Life Expectancy for each country and their trend line on different plots.

```python
plt.figure(figsize = (15,15))
plt.subplot(3, 2, 1)
sns.regplot(data = chile_data, x = 'Year', y = 'LE', color = list(country_palette.values())[0], marker = 'o')
plt.xticks(rotation = 45)
plt.ylabel('Life expectancy at birth in years')
plt.title('Chile Life Expectancy by Year')
plt.subplots_adjust(hspace =0.5)

plt.subplot(3, 2, 2)
sns.regplot(data = china_data, x = 'Year', y = 'LE', color = list(country_palette.values())[1], marker = 'o')
plt.xticks(rotation = 45)
plt.ylabel('Life expectancy at birth in years')
plt.title('China Life Expectancy by Year')

plt.subplot(3, 2, 3)
sns.regplot(data = germany_data, x = 'Year', y = 'LE', color = list(country_palette.values())[2], marker = 'o')
plt.xticks(rotation = 45)
plt.ylabel('Life expectancy at birth in years')
plt.title('Germany Life Expectancy by Year')

plt.subplot(3, 2, 4)
sns.regplot(data = mexico_data, x = 'Year', y = 'LE', color = list(country_palette.values())[3], marker = 'o')
plt.xticks(rotation = 45)
plt.ylabel('Life expectancy at birth in years')
plt.title('Mexico Life Expectancy by Year')

plt.subplot(3, 2, 5)
sns.regplot(data = usa_data, x = 'Year', y = 'LE', color = list(country_palette.values())[4], marker = 'o')
plt.xticks(rotation = 45)
plt.ylabel('Life expectancy at birth in years')
plt.title('USA Life Expectancy by Year')

plt.subplot(3, 2, 6)
sns.regplot(data = zimbabwe_data, x = 'Year', y = 'LE', color = list(country_palette.values())[5], marker = 'o')
plt.xticks(rotation = 45)
plt.ylabel('Life expectancy at birth in years')
plt.title('Zimbabwe Life Expectancy by Year')
plt.show()
```

We can confirm from the charts above that the Life Expectancy has increased over time for each individual country. <br>
We mentioned earlier that there was a recession around 2008 and that most countries experienced a dip in GDP. By observing these plots, you'll notice that only two nations also experienced a dip in life expectancy between 2008 and 2009, Chile and Mexico.

Now, let's take a look at Life Expectancy vs. GDP to see if there is any correlation between the two variables.

```python
plt.figure(figsize = (12,8))
sns.scatterplot(data = data, x = 'LE', y = 'GDP', hue = 'Country', palette= country_palette)
plt.title('Life Expectancy at Birth in Years vs. GDP')
plt.xlabel('Life expectancy at birth in years')
plt.show()
```

We see from the above chart that the plots are clustered by country. <br>Let's plot each country on it's own graph since the GDP and life expectancy varies from country to country.

```python
plt.figure(figsize = (12,20))
plt.subplot(3, 2, 1)
sns.scatterplot(data = chile_data, x = 'LE', y = 'GDP', color = list(country_palette.values())[0])
plt.xticks(rotation = 45)
plt.xlabel('Life expectancy at birth in years')
plt.title('Chile Life Expectancy vs GDP')
plt.subplots_adjust(hspace =0.35)

plt.subplot(3, 2, 2)
sns.scatterplot(data = china_data, x = 'LE', y = 'GDP', color = list(country_palette.values())[1])
plt.xticks(rotation = 45)
plt.xlabel('Life expectancy at birth in years')
plt.title('China Life Expectancy vs GDP')

plt.subplot(3, 2, 3)
sns.scatterplot(data = germany_data, x = 'LE', y = 'GDP', color = list(country_palette.values())[2])
plt.xticks(rotation = 45)
plt.xlabel('Life expectancy at birth in years')
plt.title('Germany Life Expectancy vs GDP')

plt.subplot(3, 2, 4)
sns.scatterplot(data = mexico_data, x = 'LE', y = 'GDP', color = list(country_palette.values())[3])
plt.xticks(rotation = 45)
plt.xlabel('Life expectancy at birth in years')
plt.title('Mexico Life Expectancy vs GDP')

plt.subplot(3, 2, 5)
sns.scatterplot(data = usa_data, x = 'LE', y = 'GDP', color = list(country_palette.values())[4])
plt.xticks(rotation = 45)
plt.xlabel('Life expectancy at birth in years')
plt.title('USA Life Expectancy vs GDP')

plt.subplot(3, 2, 6)
sns.scatterplot(data = zimbabwe_data, x = 'LE', y = 'GDP', color = list(country_palette.values())[5])
plt.xticks(rotation = 45)
plt.xlabel('Life expectancy at birth in years')
plt.title('Zimbabwe Life Expectancy vs GDP')
plt.show()
```

It looks like that among each country, there is a strong positive correlation between Life Expectancy and GDP. In other words, as the GDP increases, the life expectancy in the country also increases.
